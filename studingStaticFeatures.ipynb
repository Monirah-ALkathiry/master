**Bert tokenezation**

```python
num_classes = 2

#BertTokenizer.tokenize() tokenizes sequences into word-pieces.
bert_tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
```

```python
#This function used toconvert the text to features that can hnddeled by the bert classifier

def convert_text_to_feature(Data):
  return bert_tokenizer.encode_plus(Data, 
                add_special_tokens = True,     # add [CLS], [SEP]
                max_length = 512,              # max length of the text that can go to BERT
                padding='max_length',
                truncation=True,
                return_attention_mask = True,  # add attention mask to not focus on pad tokens
              )
```

```python
# map to the expected input to TFBertForSequenceClassification
def map_text_to_dict(input_ids, attention_masks, token_type_ids, label):
  return {
      "input_ids": input_ids,
      "token_type_ids": token_type_ids,
      "attention_mask": attention_masks,
  }, label
```

```python
def encode_Dataset(ds):
  # prepare list, so that we can build up final TensorFlow dataset from slices.
  input_ids_list = []
  token_type_ids_list = []
  attention_mask_list = []
  label_list = []
  for review, label in ds:
    bert_input = convert_text_to_feature(review)
    input_ids_list.append(bert_input['input_ids'])
    token_type_ids_list.append(bert_input['token_type_ids'])
    attention_mask_list.append(bert_input['attention_mask'])
    label_list.append([label])

  return tf.data.Dataset.from_tensor_slices((input_ids_list, attention_mask_list, token_type_ids_list, label_list)).map(map_text_to_dict)
```

```python
# hyper-parameters
batch_size = 8

#batch sizes: 8, 16, 32, 64, 128
#Reduce your Batchsize of datagen.flow (by default set 32 so you have to set 8/16/24 )
```

```python
# train dataset
ds_train = zip(Combine, train_labels)
ds_test = zip(val_Combine, val_labels)

ds_train_encoded = encode_Dataset(ds_train).shuffle(len(Combine)).batch(batch_size)
ds_test_encoded = encode_Dataset(ds_test).batch(batch_size)

val_test = zip( Prediction_Combine, prediction_labels)
val_test_encoded = encode_Dataset(val_test).batch(batch_size)
```
